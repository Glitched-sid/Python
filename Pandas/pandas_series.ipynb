Series is on dimentional dataset which can store one feature

```python
import numpy as np
import pandas as pd
```

```python
Series = pd.Series(np.random.randn(4))
S = pd.Series([2, 3, 4,"Hello",1,2])
print(Series)
print(S)
print(Series.index)
```

```python
dict = {'m':1,'y':2,'x':3}
Series = pd.Series(dict)
print(Series)

s = pd.Series(dict,index=['a','b','c'])
print(s)
```

```python
Series = pd.Series(np.random.randn(4),index=['a','b','c','d'])
print(Series)
print(Series.index)
```

```python
print(Series[1])
```

```python
print('creating series from scaler value')
scl = pd.Series(5, index=['a', 'b', 'c'])
print(scl)
```

```python
Series = pd.Series(np.random.randn(4),index=['a','b','c','d'])
print(Series)
print('\n mean:',Series.mean())#Calculate mean
print('\n median:',Series.median())#Calculate median
print('\n max:',Series.max())#Calculate standard deviation
print('\n min:',Series.min())#Calculate minimum
print('\n std:',Series.std())#Calculate standard deviation
print('\n sum:',Series.sum())#Calculate sum

# Series2 = pd.Series(np['a','b','c','d'])
```
