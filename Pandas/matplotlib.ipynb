```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9] #x-data
y= [2,8,4,16,10,12,6,12,8] #y-data
plt.plot(x,y) #plotting the graph using x-data and y-data
```

```python
plt.plot(x,y,color = 'black',linestyle = 'dotted') #linestyle gives the style to line #pattern
plt.plot(x,y,color = 'black',linewidth = '2') #lineswidth gives the style to line #size
plt.plot(x,y,color = 'black',marker = '*') #linestyle gives the style to line #size
plt.plot(x,y,color = 'black',markersize = 20) #linestyle gives the style to line #size

plt.title('Number of Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show() #It is used to display the graph.
```

Bargraph

```python
x = ['Python','C++','C','Java']
y = [87,62,50,80]
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y, color = "lime", width = 0.4, align = 'center', alpha = 0.5, edgecolor = 'red') #alpha - transparency
plt.title('Programming language vs popularity')
plt.xlabel("language")
plt.ylabel("popularity")
plt.show()
```

```python
plt.barh(x,y, color = 'orange', height = 0.4, label = 'popularity')
plt.title('Programming language vs popularity')
plt.xlabel("language")
plt.ylabel("popularity")
plt.legend()
plt.show()
```

pie

```python
x = [10,20,30,40,15]
l = ['C','C++','Python','Java','R']
```

```python
plt.pie(x,labels = l)
plt.show()
```

```python
ex = [0.3,0.0,0.0,0.5,0.0]
c = ['red','yellow','blue','green','orange']
plt.pie(x,labels = l,explode = ex,colors = c, autopct = "%0.1f%%", shadow = True)
plt.show()
```

```python
ex = [0.3,0.0,0.0,0.5,0.0]
c = ['red','yellow','blue','green','orange']
plt.pie(x,labels = l,explode = ex,colors = c, autopct = "%0.1f%%", shadow = True, radius = 0.5)
plt.show()
```

```python
ex = [0.3,0.0,0.0,0.5,0.0]
c = ['red','yellow','blue','green','orange']
plt.pie(x,labels = l,explode = ex,colors = c, autopct = "%0.1f%%", shadow = True, startangle= 180)
plt.show()
```

```python
x = [10,20,30,40,]
x1 = [60,50,30,20]
y = ['C','C++','Java','Python']
c = ['r','y','g','b','m']
plt.pie(x,labels = y,radius = 1.5)
plt.pie(x1,radius = 0.7, colors = c)
plt.show()
```
