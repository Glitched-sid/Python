```python

df = sns.load_dataset('penguins')
```

Heatmap()

```python
# calculate correlation matrix
correlation_matrix = df.corr(numeric_only = True)
# Heatmap
sns.heatmap(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,annot = True ,fmt = '.3f')
```

```python
sns.heatmap(correlation_matrix,annot = True ,linewidths = 2,linecolor='r')
```

```python
sns.heatmap(correlation_matrix,annot = True ,linewidths = 2,cmap = 'jet')
# cmap = viridis, summer,winter, jet
```
